```python
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.DiGraph()

graph.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F'])
graph.add_edges_from([('A', 'B'), ('A', 'C'), ('A', 'D'), ('C', 'E'), ('C', 'F'), ('D', 'F')])

pos = nx.spring_layout(graph)
nx.draw(graph,pos,with_labels=True)
plt.show()
```

```python
# Degree of nodes

print("Degree:", graph.degree())
print("In-degree:",graph.in_degree())
print("Out-degree:",graph.out_degree())
```

```python
# BFS

def bfs(graph, start):
  visited = set()
  queue = [start]
  ans = []
  while queue:
    current = queue.pop(0)
    if current not in visited:
      visited.add(current)
      ans.append(current)
      queue.extend(graph.successors(current))
  return ans

print("BFS:",bfs(graph, 'A'))
```

```python
# DFS
def dfs(graph, start):
  visited = set()
  stack = [start]
  ans = []
  while stack:
    node = stack.pop()
    if node not in visited:
      visited.add(node)
      ans.append(node)
      for neighbor in graph.neighbors(node):
        stack.append(neighbor)
  return ans

print("DFS:",dfs(graph, 'A'))
```

```python
# Density

density = nx.density(graph)
print("Density:", density)
```

```python
# Reciprocity

reciprocity = nx.reciprocity(graph)
print("Reciprocity:",reciprocity)
```
